```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\kashi\Downloads\Uber Request Data.csv")
df
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df['request_timestamp'] = pd.to_datetime(df['request_timestamp'], errors='coerce', dayfirst=True)
df['drop_timestamp'] = pd.to_datetime(df['drop_timestamp'], errors='coerce', dayfirst=True)
df
```

```python
# DateTime (for sorting/grouping)
df['request_datetime'] = df['request_timestamp']
df['drop_datetime'] = df['drop_timestamp']
df['request_date'] = df['request_timestamp'].dt.date
df['drop_date'] = df['drop_timestamp'].dt.date
df
```

```python
# 12-hour format
df['request_time'] = df['request_timestamp'].dt.strftime('%I:%M:%S %p')
df['drop_time'] = df['drop_timestamp'].dt.strftime('%I:%M:%S %p')
df
```

```python
# for completed trips
df['trip_duration_min'] = (df['drop_timestamp'] - df['request_timestamp']).dt.total_seconds() / 60
df
```

```python
df = df.drop(columns=['request_timestamp', 'drop_timestamp'], errors='ignore')
df
```

```python
df.to_excel('Uber_Request_Cleaned.xlsx', index=False)
df
```
